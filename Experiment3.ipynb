## **Experiment 3: Histogram Analysis**
Write a program to compute and plot the histogram of a grayscale image and perform histogram
equalization.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os

#  Load the image and verify
img = cv2.imread('/rit.jpeg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    #  Perform Histogram Equalization
    img_equalized = cv2.equalizeHist(img)

    #  Create the visualization plot
    plt.figure(figsize=(16, 8))

    # --- TOP LEFT: Original Image ---
    plt.subplot(2, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # --- TOP RIGHT: Original Histogram ---
    plt.subplot(2, 2, 2)
    plt.hist(img.ravel(), 256, [0, 256], color='black')
    plt.title('Original Histogram')
    plt.xlim([0, 256])

    # --- BOTTOM LEFT: Equalized Image ---
    plt.subplot(2, 2, 3)
    plt.imshow(img_equalized, cmap='gray')
    plt.title('Equalized Image')
    plt.axis('off')

    # --- BOTTOM RIGHT: Equalized Histogram ---
    plt.subplot(2, 2, 4)
    plt.hist(img_equalized.ravel(), 256, [0, 256], color='black')
    plt.title('Equalized Histogram')
    plt.xlim([0, 256])

    plt.tight_layout()
    plt.show()
```
